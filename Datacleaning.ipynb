```python
import pandas as pd
clean_frame_pred = pd.read_csv("cleaned_dataset_pred.csv")
```

```python
import numpy as np
clean_frame_pred['total'] = np.ones(len(clean_frame_pred['y']))
```

```python
clean_frame_pred.columns
```

```python
sub_pred = clean_frame_pred[['DAUID','conn_type','x','y','Y','distance_to_closest_city','closest_PC_dist',
                             'DA_POP','total','PRNAME']]
f = {'x':['mean'], 'y':['mean'],'Y':['sum'],
     'distance_to_closest_city':['mean'],'closest_PC_dist':['mean'],'DA_POP':['mean'],'total':['sum']}
grouped_df = sub_pred.groupby(by = ['DAUID','PRNAME']).agg(f)
grouped_df
```

```python
grouped_df.columns = ['x','y','Y','distance_to_closest_city','closest_PC_dist','DA_POP','total']
```

```python
mode_df= sub_pred.groupby(['DAUID','PRNAME'])['conn_type'].agg(pd.Series.mode).to_frame()
```

```python
final_clean_data_pred  = grouped_df.join(mode_df, on = ['DAUID','PRNAME'])
```

```python
final_clean_data_pred = final_clean_data_pred.reset_index(level=1)
```

```python
final_clean_data_pred.to_csv('final_clean_grouped_pred.csv')
```
